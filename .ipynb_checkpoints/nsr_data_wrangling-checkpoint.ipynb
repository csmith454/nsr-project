#### NSR Project Data Wrangling

```python
# import necessary packages
import pandas as pd
import os
```

```python
# open csv files as pandas dataframes 

start_y = 2011
max_y = 2018
curr_y = start_y

path = '~/desktop/nsr-project/NSRData_CSV/'
data = {}

while curr_y <= max_y: 
    file_name = 'nsr_' + str(curr_y) + '.csv'
    df = pd.read_csv(os.path.join(path,file_name))
    data[curr_y] = df
    curr_y +=1
```

#### Clean up each dataframe

```python
# 2011

df = data[2011]
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# make row index 0 into column headers
df.columns = df.iloc[0]
df = df[1:]
df.reset_index(drop=True, inplace=True)
df.rename(columns=str)

df.info()
```

```python
# change column header names
df.columns = ['index','index2','vessel name','type','flag','cargo','port of loading','port of destination']
df.columns
```

```python
df = df.drop(['index','index2'],axis=1)
```

```python
df.head()
```

```python
df = df.replace('\n',' ',regex=True)
```
